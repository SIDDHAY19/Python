```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        if not self.head:
            self.head = Node(data)
        else:
            cur = self.head
            while cur.next:
                cur = cur.next
            cur.next = Node(data)

    def display(self):
        cur = self.head
        while cur:
            print(cur.data)
            cur = cur.next

class Stack:
    def __init__(self):
        self.stack = []

    def push(self, data):
        self.stack.append(data)

    def pop(self):
        if len(self.stack) < 1:
            return None
        return self.stack.pop()

    def display(self):
        for item in reversed(self.stack):
            print(item)

class Queue:
    def __init__(self):
        self.queue = []

    def enqueue(self, data):
        self.queue.append(data)

    def dequeue(self):
        if len(self.queue) < 1:
            return None
        return self.queue.pop(0)

    def display(self):
        for item in self.queue:
            print(item)

def main():
    data_structures = {
        '1': LinkedList(),
        '2': Stack(),
        '3': Queue()
    }

    while True:
        print("Select a data structure:")
        print("1. Linked List")
        print("2. Stack")
        print("3. Queue")
        print("4. View data in data structures")
        print("5. Exit")
        choice = input("Enter your choice (1-5): ")

        if choice == '5':
            print("Exiting...")
            break

        if choice == '4':
            for key, ds in data_structures.items():
                print(f"Data in data structure {key}:")
                ds.display()
            continue

        if choice in data_structures:
            ds = data_structures[choice]
            while True:
                if isinstance(ds, LinkedList):
                    print("1. Insert data")
                    print("2. Exit")
                    operation = input("Enter operation number: ")

                    if operation == "1":
                        data = input("Enter data to insert: ")
                        ds.insert(data)
                    elif operation == "2":
                        break
                    else:
                        print("Invalid operation. Please try again.")

                elif isinstance(ds, Stack):
                    print("1. Push data")
                    print("2. Pop data")
                    print("3. Exit")
                    operation = input("Enter operation number: ")

                    if operation == "1":
                        data = input("Enter data to push: ")
                        ds.push(data)
                    elif operation == "2":
                        popped_data = ds.pop()
                        if popped_data:
                            print("Popped data:", popped_data)
                        else:
                            print("Stack is empty.")
                    elif operation == "3":
                        break
                    else:
                        print("Invalid operation. Please try again.")

                elif isinstance(ds, Queue):
                    print("1. Enqueue data")
                    print("2. Dequeue data")
                    print("3. Exit")
                    operation = input("Enter operation number: ")

                    if operation == "1":
                        data = input("Enter data to enqueue: ")
                        ds.enqueue(data)
                    elif operation == "2":
                        dequeued_data = ds.dequeue()
                        if dequeued_data:
                            print("Dequeued data:", dequeued_data)
                        else:
                            print("Queue is empty.")
                    elif operation == "3":
                        break
                    else:
                        print("Invalid operation. Please try again.")

                else:
                    print("Invalid data structure.")

if __name__ == "__main__":
    main()
```
